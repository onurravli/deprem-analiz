```python
from matplotlib import pyplot as ax
import requests
```

```python
class earthquake:
    def __init__(self, magnitude, depth, location, time):
        self.magnitude = magnitude
        self.depth = depth
        self.location = location
        self.time = time

    def getMagnitude(self):
        return self.magnitude

    def getDepth(self):
        return self.depth

    def getLocation(self):
        return self.location

    def getTime(self):
        return self.time

    def __str__(self):
        return "Magnitude: " + str(self.magnitude) + " Depth: " + str(self.depth) + " Location: " + str(self.location) + " Time: " + str(self.time)
```

```python
def fetch():
    url = 'https://api.orhanaydogdu.com.tr/deprem/kandilli/live'
    response = requests.get(url)
    return response.json()
```

```python
data = fetch()['result']
```

```python
eqs = []

for eq in data:
    dummy = earthquake(eq['mag'], eq['depth'], eq['lokasyon'], eq['date'])
    eqs.append(dummy)
```

```python
x = []
y = []

for eq in eqs:
    x.append(eq.getMagnitude())
    y.append(eq.getDepth())

ax.plot(x, y, 'ro')
```
